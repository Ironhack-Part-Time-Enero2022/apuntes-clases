```python
fig, ax = plt.subplots(5, 1, figsize=(30,20))

for i in range(len(df_numericas.columns)):
    sns.boxplot(x=df_numericas.columns[i], data=df_numericas, ax=ax[i])
plt.show()
```

```python
def detectar_outliers(lista_columnas, dataframe): 
    
    dicc_indices = {} # creamos un diccionario donde almacenaremos 칤ndices de los outliers
    
    # iteramos por la lista de las columnas num칠ricas de nuestro dataframe
    for col in lista_columnas:
        
        #calculamos los cuartiles Q1 y Q3
        Q1 = np.nanpercentile(df[col], 25)
        Q3 = np.nanpercentile(df[col], 75)
        
        # calculamos el rango intercuartil
        IQR = Q3 - Q1
        
        # calculamos los l칤mites
        outlier_step = 1.5 * IQR
        
        # filtramos nuestro dataframe para indentificar los outliers
        outliers_data = dataframe[(dataframe[col] < Q1 - outlier_step) | (dataframe[col] > Q3 + outlier_step)]
        
        
        if outliers_data.shape[0] > 0: # chequeamos si nuestro dataframe tiene alguna fila. 
        
            dicc_indices[col] = (list(outliers_data.index)) # si tiene fila es que hay outliers y por lo tanto lo a침adimos a nuestro diccionario
        

    
    return dicc_indices 
```

```python
lista_columnas = df_numericas.columns
lista_columnas
```

Antes de empezar con los modelos voy a explicar brevemente las m칠tricas que podemos usar para evaluar su bondad. Es decir, la fiabilidad de las predicciones del modelo: 

- `Matriz de confusi칩n`: Una matriz de confusi칩n es un resumen de los resultados de las predicciones en un problema de clasificaci칩n.

    El n칰mero de predicciones correctas e incorrectas se resume con valores de recuento y se desglosa por cada clase, es decir, la matriz de confusi칩n muestra las formas en que el modelo se confunde cuando hace predicciones.
    
    - `Verdadero positivo`:
        Interpretaci 칩n: Predijiste positivo y es cierto.
        Predijiste que una mujer est치 embarazada y realmente lo est치.
    - `Verdadero Negativo`:
        Interpretaci칩n: Predijiste negativo y es cierto.
        Predijiste que un hombre no est치 embarazado y en realidad no lo est치.
    - `Falso positivo`: Error de tipo 1
        Interpretaci칩n: Predijiste positivo y es falso.
        Predijiste que un hombre est치 embarazado pero en realidad no lo est치.
    - `Falso negativo`: Error de tipo 2
        Interpretaci칩n: Predijiste negativo y es falso.
        Predijiste que una mujer no est치 embarazada pero en realidad lo est치.

    丘멆잺 **NOTA** El foco de este tipo de errores depende de la pregunta que nos estemos haciendo. Es decir, en un ejemplo de predici칩n de cancer lo ideal es que tanto los verdaderos positivos y verdaderos negativos alcancen los valores m치s altos. Mientras que deber칤amos obtener mayor n칰mero de falsos positivos que negativos. Es decir, es mejor predecir que vas a tener iun infarto y no lo vas a tener, que predecir que no tienes un infarto y lo tienes. 
    
    Sin embargo, esta m칠trica es compleja de interpretar y llevar칤a a un estudio detallado dependiendo del caso de estudio. 


- `Accuracy`: es lo cerca que est치 un valor de su valor real. Un ejemplo es lo cerca que est치 una flecha del centro de la diana.


- `Precisi칩n` : es lo repetible que es una medici칩n. Un ejemplo es lo cerca que est치 una segunda flecha de la primera (independientemente de que cualquiera de las dos est칠 cerca del centro de la diana)


- `Recall`: la capacidad de un modelo para encontrar todos los casos relevantes dentro de un conjunto de datos. Es decir, definimos el "recall" como el n칰mero de verdaderos positivos dividido por el n칰mero de verdaderos positivos m치s el n칰mero de falsos negativos.

- `F1`: es la media de la precisi칩n y el recuerdo, teniendo en cuenta ambas m칠tricas en la siguiente ecuaci칩n:

```python
# vamos a hacernos una funci칩n
def metricas(clases_reales_test, clases_predichas_test, clases_reales_train, clases_predichas_train, modelo):
    
    # para el test
    accuracy_test = accuracy_score(clases_reales_test, clases_predichas_test)
    precision_test = precision_score(clases_reales_test, clases_predichas_test)
    recall_test = recall_score(clases_reales_test, clases_predichas_test)
    f1_test = f1_score(clases_reales_test, clases_predichas_test)
    
    # para el train
    accuracy_train = accuracy_score(clases_reales_train, clases_predichas_train)
    precision_train = precision_score(clases_reales_train, clases_predichas_train)
    recall_train = recall_score(clases_reales_train, clases_predichas_train)
    f1_train = f1_score(clases_reales_train, clases_predichas_train)
    

    
    df = pd.DataFrame({"accuracy": [accuracy_test, accuracy_train], 
                       "precision": [precision_test, precision_train],
                       "recall": [recall_test, recall_train], 
                       "f1": [f1_test, f1_train],
                       "set": ["test", "train"]})
    
    df["modelo"] = modelo
    return df
```

## KNeighborsClassifier 

El algoritmo KNN utiliza la "similitud de caracter칤sticas" para predecir los valores de cualquier punto de datos nuevo. Esto significa que al nuevo punto se le asigna un valor basado en su parecido con los puntos del conjunto de entrenamiento. 

A continuaci칩n se explica el algoritmo paso a paso:

1. En primer lugar, se calcula la distancia entre el nuevo punto y cada punto de entrenamiento.

2. Se seleccionan los k puntos de datos m치s cercanos (en funci칩n de la distancia). En este ejemplo, se seleccionar치n los puntos 1, 5, 6 si el valor de k es 3. M치s adelante exploraremos el m칠todo para seleccionar el valor correcto de k.

El algoritmo k-pr칩ximos (KNN) es un algoritmo de aprendizaje autom치tico supervisado sencillo y f치cil de implementar que puede utilizarse para resolver problemas de clasificaci칩n y regresi칩n. 

Lo primero que tenemos que hacer es conocer cual es el n칰mero 칩ptimo de vecinos que tenemos que establecer en nuestro modelo. 

Para eso realizamos un `for loop` testanto diferente n칰mero de vecinos y calculando el `accuracy` del modelo para conocer cual es el mejor

```python
knn_neighbors = range(1,21)
knn_scores = []

for k in knn_neighbors:
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score=cross_val_score(knn_classifier,X,y,cv=10)
    knn_scores.append(score.mean())
```

```python
knn = pd.DataFrame(knn_scores, knn_neighbors).reset_index()
knn.columns = ["number_neighbors", "score"]
knn.sort_values(by = "score", ascending = False).head(3)
```

游늷 El mejor n칰mero de vecinos para nuestros datos es `7`. Por lo tanto, volveremos a entrenar el modelo con los dataos para ese n칰mero de vecinos y evaluar el modelo
