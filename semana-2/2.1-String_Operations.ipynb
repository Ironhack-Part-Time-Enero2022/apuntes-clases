# Métodos de los strings

Los *strings* van a tener sus métodos propios. Veamos los principales:

````python
string = 'Python and data science and machine learning'
````

`upper`: devuelve el *srting* con todos sus caracteres en mayúscula

````python
string.upper()
````

````python
string_upper = string.upper()
````

`isupper`: chequea si todos los caracteres son mayúscula

````python
string.isupper()
````

````python
string_upper.isupper()
````

`lower`: devuelve la cadena con todos sus caracteres en minúscula

````python
string.lower()
````

````python
string_upper = string.lower()
````

`islower`: chequea si todos los caracteres son minúsculas

````python
string.isupper()
````

````python
string_upper.islower()
````

`capitalize`: devuelve la cadena con sus primer carácter en mayúsucula 

````python
string.capitalize()
````

`count`: devuelve el número de veces que aparece un *substring* en el *string*

````python
string.count("machine")
````

````python
string.count("and")
````

````python
string.count("e")
````

`title`: devuelve el *string* con el primer carácter de cada palabra en mayúscula

````python
string.title()
````

`find`: devuelve el índice de la primera aparición del carácter dado en una cadena

````python
string.find("o")
````

````python
# También podemos encontrar la segunda u otras apariciones de un personaje.
string.find("o", 2)
````

````python
string.find("la")
````

````python
string.find("la")
````

`isnumeric` : chequea si todos los caracteres son números

````python
string.isnumeric()
````

`isalpha`: cheaquea si todos los caracteres están en el alfabeto

````python
string.isalpha()
````

````python
# lo anterior 👆🏼 nos da false porque hay espacios!!! 

string2 = "hola"
string2.isalpha()
````

`isalnum`: chequea si los caracteres son alfanuméricos, es decir, números o letras. 

````python
string.isalnum()
````

`startswith`: devuelve True si el *string* empieza con el *substring* que le pasemos

````python
string.startswith("h")
````

````python
# también les podemos pasar una secuencia de caracteres
string.startswith("ho")
````

````python
# ojo! Es case-sensitive
string.startswith("Ho")
````

`endswith`: devuelve True si el *string* acaba con el *substring* que le pasemos

````python
string.endswith("do")
````

````python
string2 = "   Hola mundo, estamos aprendiendo a ser programadores----"
````

`split`: separa el *string* por el símbolo que especifiquemos. Por defecto espacio.

📌 Nos devuelve una lista

⚠️🚨 **ATENCION, ELIMINARÁ EL ELEMENTO POR EL QUE SEPAREMOS** ⚠️🚨

````python
string2.split()
````

````python
string2.split(",")
````

````python
lista = string2.split(",")
lista
````

````python
string.split("la")
````

`partition`: divide una cadena en **3 partes** y devuelve una tupla que contiene estas partes

📌 Nos devuelve una tupla

⚠️🚨 **ATENCION, NO ELIMINARÁ EL ELEMENTO POR EL QUE SEPAREMOS** ⚠️🚨

````python
string.partition("data")
````

````python
string.partition("and")
````

`join`: une todos los caracteres de una cadena utilizando un caracter de unión que nosotros especifiquemos. 

📌 Nos devuelve un *string* 

````python
",".join(lista)
````

````python
"🤓".join(lista)
````

`strip`: borra todos los espacios por delante y detrás de un *string*. Por defecto espacios

````python
string2
````

````python
string2.strip()
````

````python
# podemos indicar el carácter a borrar
string2.strip('-')
````

````python
# podemos encadenar todos los métodos que queramos
string2.strip('-').strip()
````

`rstrip`: borra todos los espacio que a la derecha de nuestro *string*.  Por defecto espacio, pero al igual que el `strip` podemos indicar qué queremos eliminar

````python
string2.rstrip("-")
````

`lstrip`: borra todos los espacio que a la izquierda de nuestro string. Por defecto espacio, pero al igual que el strip podemos indicar qué queremos eliminar

````python
string2.lstrip()
````

`replace`: reemplaza un *substring* de un *string* pot otra

````python
string2.replace("o", "O")
````
