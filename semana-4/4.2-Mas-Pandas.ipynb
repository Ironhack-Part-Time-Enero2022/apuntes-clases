## `replace()`

Reemplaza ciertos valores de nuestro dataframe por otro valor que nosotros especifiquemos. 

```python
MiDataframe.replace(to_replace, value, inplace)
```

Donde:

- `to_replace`: es el valor que queremos reemplazar


- `value`: el valor nuevo que queremos incluir


- `inplace`: para que los cambios se ejecuten sobre el dataframe orginal.

````python
df.head(2)
````

## `np.where()`

La sintaxis que seguiremos en este m√©todo es: 

```python
MiDataframe.where(cond, true, false)
```

Donde: 

- `cond`: es una condici√≥n que nosotros querramos que cumpla el dataframe.


- `true`: es el valor que queremos que se reemplace en el dataframe cuando SI se cumpla la condici√≥n.


- `false`: es el valor que queremos que se reemplace en el dataframe cuando NO se cumpla la condici√≥n.



````python
df["desciption_cambiada"] = np.where(df["description"] == "CREAM HANGING HEART T-LIGHT HOLDER", 1, 2)
````

````python
# si nos fijamos en la nueva columna, tendremos valores de 1 donde se cumpla la cumpla la condici√≥n y 2 donde no. 
df.sample(10)
````

# `loc` e `iloc`

Las principales diferencias entre `loc` y `iloc` es:


- `loc` se basa en etiquetas, lo que significa que tiene que especificar filas y columnas en funci√≥n de sus etiquetas de fila y columna.


- `iloc`:  se basa en la posici√≥n de los enteros, por lo que hay que especificar las filas y las columnas por sus valores de posici√≥n de los enteros (los √≠ndices)

````python
# antes de empezar, para entender mejor esta parte vamos a poner la 
## columna "Source" como √≠ndice

df.set_index(["customerid"], inplace = True)
df.head()
````

# Apply

Aplicar una funci√≥n a un DataFrame de Pandas es una de las operaciones m√°s comunes durante la manipulaci√≥n de datos.

La funci√≥n `apply` de Pandas DataFrame es una de las mejores herramientas que tenemos en pandas para hacerlo. Toma una funci√≥n como argumento y la aplica a lo largo de un eje del DataFrame.

````python
# como es un csv enoooorme voy a coger solo 5000 filas para los ejemplos de la clase:

dff = df.sample(100)
````

````python
#recordemos como es nuestro dataframe

dff.head()
````

Si nos fijamos, tenemos una columna `invoicedata` que contiene muchos datos y que vamos a querer separar en varias columnas. Vamos a ello: 

````python
# lo primero que tenemos que hacer es una funci√≥n que haga el split de los datos que queremos. 
def limpiar_fecha(x): 
    '''
    Esta funci√≥n va a extraer la fecha de un string
    args:
        x (str): una fecha en formato mes - mes - dia hora
    returns:
        un objeto de tipo datetime √∫nicamente con la fecha
    '''
    
    fecha =  datetime.datetime.strptime(x.split(" ")[0], '%Y-%m-%d')
    return fecha
````

````python
dff['fecha'] = dff['invoicedate'].apply(limpiar_fecha)
````

````python
dff.dtypes
````

````python
dff.head()
````

````python
def limpiar_hora(x): 
    hora =  datetime.datetime.strptime(x.split(" ")[0].split("."), '%Y-%m-%d')
    return hora
````

Imaginemos que queremos sacar las coordenadas como latitud y longitud de nuestros paises. Tambi√©n lo podemos hacer usando la librer√≠a `geopy`.

[Aqu√≠](https://pypi.org/project/geopy/) la documentaci√≥n a la p√°gina oficial. 

````python
# lo primero que tenemos que hacer es importar la librer√≠a
from geopy.geocoders import Nominatim

def extraer_coordenadas(pais):
    try:
        geolocator = Nominatim(user_agent="ana")
        location = geolocator.geocode(pais)
        return location.latitude, location.longitude
    except: 
        return "sorry we dont have data"
````

````python
%%time
dff["coordenadas"] = dff["country"].apply(extraer_coordenadas)
````

**TAMBI√âN PODEMOS HACER UN APPLY A MULTIPLES COLUMNAS** 

Imaginemos que queremos saber el dinero gastado por un cliente en sus compras. 
Podemos hacernos una funci√≥n a la que le pasemos dos columnas y nos devuelva una columna nueva

````python
def precio_total(precio, unidades):
    return precio * unidades
````

````python
dff["precio_total"] = dff.apply(lambda data: precio_total(data["unitprice"], data["quantity"]), axis = 1)
````

````python
dff.head()
````

# Groupby

En el an√°lisis exploratorio de datos, a menudo nos gustar√≠a analizar los datos por algunas categor√≠as. 

En Pandas, tenemos el m√©todo `groupby()` que nos permite dividir los datos en grupos separados para realizar c√°lculos para un mejor an√°lisis.



````python
dff.head(1)
````

````python
# lo primero que vamos a hacer es resetear el √≠ndice

dff.reset_index(["customerid"], inplace = True)
dff.head()
````

````python
# imaginemos que queremos saber la media de compras que hizo un cliente en fncion del producto
dff.groupby(["customerid", "description"]).mean().head()
````

Vale... y si quiero solo la media de una columna, es decir, solo me interesa el precio_total que se ha gastado el cliente: 

````python
dff.groupby(["customerid", "description"])["precio_total"].mean().head()
````

¬øPodemos almacenar estos resultados en un dataframe? Esto es pandas, claro que SI!!!!

````python
precio_cliente = pd.DataFrame(dff.groupby(["customerid", "description"])["precio_total"].median())
precio_cliente.head()
````

Ahora tenemos `customerid` y `description` como √≠ndices en nuestro dataframe. 

¬øRecordamos como los pod√≠amos volver a colocar como columnas? ü§î 

````python
precio_cliente.reset_index(["customerid", "description"],inplace = True)
precio_cliente.head()
````

‚ö†Ô∏è En este caso hemos usado la media pero podemos aplicar otras m√©tricas como: 

- count() ‚Äì n√∫mero de observaciones no nulas 


- sum() ‚Äì suma de todos los valores


- mean() ‚Äì media de los valores


- median() ‚Äì mediana de los valores


- min() ‚Äì valor m√≠nimo


- max() ‚Äì valor m√°ximo


- std() ‚Äì desviaci√≥n est√°ndar


- var() ‚Äì varianza


Pero estos resultados puede que no sean lo que nos interesen al 100% ya que podemos estar interesados en sacar varias m√©tricas. Pero la funci√≥n `agg()` de Pandas nos da la flexibilidad de realizar varios c√°lculos estad√≠sticos a la vez. As√≠ es como funciona:



Imaginemos que queremos sacar la media y la mediana para cada cliente:

````python
dff.groupby(["customerid"]).agg([np.mean,np.median])
````

¬øY si solo quiero los estad√≠sticos para una √∫nica columna? Por ejemplo `precio_total`? 

````python
clientes_media_mediana = pd.DataFrame(dff.groupby(["customerid"])["precio_total"].agg([np.mean,np.median])).reset_index()
clientes_media_mediana.head()
````

````python
from matplotlib import pyplot as plt

plt.figure(figsize=(30,15)) # para hacer nuestra figura m√°s grande. 


plot3 = sns.barplot(data=clientes_media_mediana, x="customerid", y="mean")
plot3.set_xticklabels(plot3.get_xticklabels(), rotation=30);
````
