# Tipos de datos


## Números 

| Clase   |      Descripción      |  Ejemplo |
|----------|:-------------:|------:|
|integer| Número entero con precisión fija  | 45 |
| float |  Número decimal     |2.19   |
| complejo | Parte real y parte imaginaria |   (4.5 + 3j)  |

### Integer

Los *integer* o números enteros son aquellos que no tienen decimales, tanto positivos como negativos (además del cero). En Python se representan mediante el tipo `int`

```python
edad = 27
```

```python
print(edad, type(edad))
```

```python
edad = 0
```

```python
print(edad, type(edad))
```

```python
edad = -3
```

```python
print(edad, type(edad))
```

### Float o números reales

Los números reales son los que tienen decimales. En Python se expresan como tipo `float`

```python
precio = 33.9
```

```python
print(precio, type(precio))
```

```python
precio2 = -3.6
```

```python
print(precio2, type(precio))
```

```python
precio3 = .6
```

```python
print(precio3, type(precio))
```

### Complejo

Los números complejos son aquellos que tienen parte imaginaria. En Python se expresan como `complex`

```python
complejo = 2.1 + 7.8j
```

```python
print(complejo, type(complejo))
```

## Cadenas de caracteres (Strings) 

Las cadenas de caracteres, son secuencias que contienen caracteres encerrado entre comillas.

| Clase   |      Descripción      |  Ejemplo |
|----------|:-------------:|------:|
|Cadenas cortas (str)| Caracteres encerrado entre comillas simples (') o dobles (") | "Hola" o 'Hola'|
| Cadenas largas (str) | Caracteres encerrados entre grupo comillas triples simples (''') o dobles (""") | """Clase que representa una Persona""" o  '''Clase que representa una Persona'''   |


### Cadenas cortas 

```python
presentacion = "Hola que tal?"
```

```python
print(presentacion, type(presentacion))
```

⚠️ Tenemos que tener cuidado con las comillas. Imaginad que queremos poner un apóstrofe, muy usados en ingles:

```python
# Si usamos comillas dobles no tendremos ningún problema
"pepe's bakery"
```

```python
# pero si usaramos comillas simples tendríamos un error de sintaxis
'pepe's bakery'
```

### Cadenas largas

**NOTA**: normalmente se usan para string multi-línea

```python
frase = """
En un lugar de la Mancha, de cuyo nombre no quiero acordarme, no ha mucho tiempo que vivía un hidalgo de los
de lanza en astillero, adarga antigua, rocín flaco y galgo corredor..
"""
```

```python
print(frase, type(frase))
```

## Booleanos

El tipo booleano sólo puede tener dos valores: True (verdadero) y False (falso). Estos valores son especialmente importantes para las expresiones condicionales y los bucles. 

| Clase   |      Descripción      |  Ejemplo |
|----------|:-------------:|------:|
|bool | Valor booleano falso	 | False|
| bool | Valor booleano verdadero	 | True  |


```python
bool()
```

```python
print(False, type (False))
```

```python
print(True, type (True))
```

```python
not True
```

```python
not False
```

📌 Este tipo de datos *a priori* nos pueden resultar muy simple pero son ESENCIALES. 

## None

Es usado para definir valores nulos. 



Datos interesantes sobre el `None`

- NO es lo mismo que False.
- NO es 0.
- NO es un string vacío.


Comparar `None` con cualquier cosa siempre va a devolver False excepto si lo comparas con otro `None`

```python
None
```

```python
type(None)
```

# Operadores

## Operadores aritméticos

Toman dos operandos como entrada, realiza un cálculo y devuelve el resultado. 

`+`: suma

```python
3 + 2
```

```python
"hola" + " " + "que tal"
```

`-`: resta

```python
5 - 7
```

```python
"hola" - "que tal"
```

`*`: multiplicación

```python
4 * 896
```

```python
"hola" * "que tal"
```

`/`: división

```python
789 / 6
```

```python
"hola" / "que tal"
```

`//`: division entera

```python
789 // 6
```

`%`: modulo, es resto de la división

```python
789 % 6
```

`**`: exponente

```python
2 ** 6
```

## Operadores de asignaciones

Imaginamos que tenemos una variable a la que le queremos sumar otro numero. Con los operadores anteriores lo podemos hacer


```python
r = 10
r = r + 5
print(r)
```

----
En Python tenemos los operadores de asignaciones que nos permiten realizar estas operaciones de una forma más sencilla, usando el operador `=`. 


>`+=` , `-=`, `*=`,  `/=`, `**=`, `//=`, `%=`: calcula la suma, resta,  multiplicación, división, exponente, división entera o módulo de la variable del lado izquierdo el valor del lado derecho

```python
r = 10
```

```python
r += 5
print(r)
```

## Operadores relacionales

Se emplean para comparar y establecer la relación entre dos elementos. Devuelven un valor booleano (True o False) basado en la condición. 

`==`: evalua que los valores sean iguales para varios tipos de datos

```python
5 == 3
```

```python
3 == 3 
```

```python
"avión" == 5
```

```python
"avión" == "avión"
```

```python
True == False
```

```python
type("avión") == str
```

`!=`: evalua si los valores son distintos

```python
5 != 3
```

```python
"avión" != 3
```

```python
"avión" != "avión"
```

```python
True != False
```

`>`: evalua soi el valor del lado izquierdo es mayor que el valor del lado derecho

```python
5 > 3
```

```python
3 > 5
```

También podemos combinarlo con el operador =. `>=` que evalua si el valor del lado izquierdo es mayor o igual que el del lado derecho

```python
3 >= 5
```

```python
3 >= 3
```

`<`: evalua si el valor del lado izquierdo es menor que el lado derecho

```python
5 < 3
```

```python
3 < 5
```

De la misma que en el operador anterior tenemos el `<=` que evalua si el valor del lado izquierdo es menor o igual que el valor del lado derecho. 

```python
3 <= 3
```

```python
5 <= 3
```

## Operadores lógicos

### Operador lógico `and`

Se tienen que cumplir todas las condiciones

```python
numero1 = 5
numero2 = 10
```

```python
eleccion = int(input('Elige un número '))
```

```python
eleccion > numero1 and eleccion < numero2 
```

### Operador lógico `or`

Se puede cumplir una condición u otra

```python
eleccion > numero1 or eleccion < numero2 
```

## Otros operadores

### Operador de afiliciación 

Son operadores utilizados para chequear la pertenencia de un valor. Comprueba la pertenencia a una secuencia, como cadenas, listas o tuplas. 

```python
lista = range(1,10)
```

```python
x = 5
y = 20
```

```python
x in lista
```

```python
y in lista
```

```python
x not in lista
```

```python
y not in lista
```

### Operador de identidad

Determinan si un valor es de una determinada clase o tipo. Tambien suelen utilizarse para determinar el tipo de datos que contiene una determinada variable. 

```python
z = 3
w = 4
```

```python
z is w
```

```python
z is not w
```

# Conversiones 

Las principales conversiones que podemos hacer en python son: 

- Convertir a *string*: str()


- Convertir a entero: int()


- Convertir a *float* (números decimales): float()


- Convertir a booleano: bool()

## Conversión de entero, decimal, y booleano a string 

```python
num1 = 23
num2 = 3.7
bool_ = True
```

```python
str(num1)
```

```python
str(num2)
```

```python
str(bool_)
```

## Conversión de entero y string a *float*

```python
num3 = 12
stri1 = "13.9"
```

```python
float(num3)
```

```python
float(stri1)
```

## Conversión de entero, decimal, y string a booleano

```python
num4 = 0
num5 = 10
num6 = 3.19
stri2 = ""
stri3 = "Hola"
```

```python
bool(num4)
```

```python
bool(num5)
```

```python
bool(num6)
```

```python
bool(stri2)
```

```python
bool(stri3)
```

📌 De estos ejemplos concluimos que los valores vacíos o nulos, cuando se les aplica la función bool(), se convierten en el False booleano. Por el contrario, cualquier otro valor equivale a True.
